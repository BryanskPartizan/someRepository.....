```python
import random


propability = float(float(input('Введите вероятность успеха в %: ')) / 100)
random_seed = random.random()

if random_seed <= propability:
    print(f'value: {random_seed}, propability: {propability}')
    print('SUCCESS')
else:
    print(f'value: {random_seed}, propability: {propability}')
    print('FAILURE')
    
#раскрытие подразделения: 10%
#неудача легкого: 50%
#неудача среднего: 60%
#неудача крупного: 65%
#убийство НПС: 20%
#убийство игрока: 1%

```
